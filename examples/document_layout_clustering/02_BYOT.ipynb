```python
import cv2
import os
import json
import hashlib
import uuid
import shutil
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
# Define the directory containing images
image_dir = "/data/user/nevil/results/clustering/hospitalwise/images/by_name/FortuneUrology"
output_dir = "/data/user/nevil/results/clustering/hospitalwise/images/by_name/FortuneUrology/auto_tagging"
config_path = "/data/user/nevil/results/clustering/hospitalwise/images/by_name/FortuneUrology/config/annotations.json"
```

```python
# Load the JSON configuration
with open(config_path, 'r') as config_file:
    config = json.load(config_file)

# Iterate through each image in the directory
for image_filename in os.listdir(image_dir):
    if image_filename.endswith(('.jpg', '.png', '.jpeg')):  # Adjust file extensions as needed
        image_path = os.path.join(image_dir, image_filename)
        image = cv2.imread(image_path)

        # Get the image's width and height
        image_width, image_height = image.shape[1], image.shape[0]

        # Iterate through the bounding box configurations in the JSON
        for bbox_config in config['annotations'][0]['result']:
            x = int(bbox_config['value']['x'] * image_width)
            y = int(bbox_config['value']['y'] * image_height)
            width = int(bbox_config['value']['width'] * image_width)
            height = int(bbox_config['value']['height'] * image_height)

            label = bbox_config['value']['rectanglelabels'][0]

            # Draw bounding box on the image
            cv2.rectangle(image, (x, y), (x + width, y + height), (0, 255, 0), 2)

            # Add label text
            cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Save the modified image with bounding boxes
        output_image_path = os.path.join(output_dir, image_filename)  # Set the output directory
        
        cv2.imwrite(output_image_path, image)
        cv2.imshow('image', image)
        print(f'Bounding boxes applied to {image_filename} and saved as {output_image_path}')

print('Processing complete.')
```

```python
def generate_unique_id(string1, string2=None):
    """
    Generate a unique ID based on one or two strings.

    Args:
    - string1 (str): The first string to be included in the unique ID.
    - string2 (str, optional): The second string to be included in the unique ID (optional).

    Returns:
    - str: The generated unique ID.
    """
    
    if string1 == None or isinstance(string1, float): 
        string1 = 'NA'
        
    # Create a hash object using SHA-256
    hasher = hashlib.sha256()

    # Update the hash object with the first string
    hasher.update(string1.encode())

    # If a second string is provided, update the hash object with it
    if string2 is not None and isinstance(string1, str):
        hasher.update(string2.encode())

    # Get the hexadecimal representation of the hash
    hash_value = hasher.hexdigest()

    # Generate a unique ID-like format from the hash value
    unique_id = hash_value[:8] + '-' + hash_value[8:12] + '-' + hash_value[12:16] + '-' + hash_value[16:20] + '-' + hash_value[20:32]

    return unique_id

# Example usage:
string1 = "Nemcare Hospital"
string2 = None
unique_id_result = generate_unique_id(string1, string2)
print("Generated Unique ID:", unique_id_result)
```

```python
data_df['hospital_ID'] = data_df['raw_hospital_name'].apply(generate_unique_id)
data_df['hospital_adress_ID'] = data_df.apply(lambda x: generate_unique_id(x['raw_hospital_name'], x['address']), axis=1)
data_df
```
